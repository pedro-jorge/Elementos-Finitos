```python
import googleapiclient.discovery as d
import re

api_service_name = "youtube"
api_version = "v3"
api_key = "AIzaSyAn3hjX3AyYyhPCp4gdRqxoR3TwtsSj-Yo"

youtube = d.build(
    api_service_name,
    api_version,
    developerKey=api_key
)

video_url = "https://www.youtube.com/watch?v=X4wge_avH2s"
max_results = 50
```

```python
def get_comment(response):
    return response["snippet"]["topLevelComment"]["snippet"]["textOriginal"]


def print_replies(response):
    try:
        n = len(response["replies"]["comments"])
        
        for i in range(n):
            print(response["replies"]["comments"][i]["snippet"]["textOriginal"])
    except:
        pass
```

```python
def get_video_id(url):
    # Regular expression pattern to match a YouTube video ID
    pattern = r'(?:v=|\/)([0-9A-Za-z_-]{11}).*'
    
    # Search for the pattern in the URL
    match = re.search(pattern, url)
    
    # If a match is found, return the video ID
    if match:
        return match.group(1)
    else:
        return None
    
video_id = get_video_id(video_url)
```

```python
request = youtube.commentThreads().list(
    part="id,replies,snippet",
    videoId=video_id,
    maxResults=max_results
)
response = request.execute()

for i in range(max_results):
    print(get_comment(response["items"][i]))
    print_replies(response["items"][i])
    
    print("")
    # for item in response["items"][i]["replies"]["comments"]:
    #     print(item["snippet"]["textOriginal"])
    #     print("")
```
